```python
asteroids_data = [] # Initialize empty list
target = 10000 # target given is 10000
url = f"https://api.nasa.gov/neo/rest/v1/feed?start_date=2020-01-01&end_date=2020-01-08&api_key={API_Key}" # Self url
while len(asteroids_data)<target:
    response = requests.get(url)
    data = response.json()
    details = data['near_earth_objects']
    for date, info in details.items():
        for neo in info:
             if neo.get("close_approach_data"):
                 approach_data = neo["close_approach_data"][0]
                 asteroids_data.append(dict(date_collected = datetime.strptime(date,"%Y-%m-%d"),
                 id = int(neo["id"]),
                 neo_ref_id = int(neo["neo_reference_id"]),
                 name = neo["name"],
                 absolute_magnitude_h = float(neo['absolute_magnitude_h']),
                 estimated_diameter_min_km = float(neo["estimated_diameter"]["kilometers"]["estimated_diameter_min"]),
                 estimated_diameter_max_km =  float(neo["estimated_diameter"]["kilometers"]["estimated_diameter_max"]),
                 is_potentially_hazardous_asteroid = neo ["is_potentially_hazardous_asteroid"],
                 close_approach_date = datetime.strptime(approach_data["close_approach_date"],"%Y-%m-%d"),
                 relative_velocity_kmph = float(approach_data["relative_velocity"]["kilometers_per_hour"]),
                 astronomical = float(approach_data["miss_distance"]["astronomical"]),
                 miss_distance_km = float(approach_data["miss_distance"]["kilometers"]),
                 miss_distance_lunar = float(approach_data["miss_distance"]["lunar"]),
                 orbiting_body = approach_data["orbiting_body"]))
                 if len(asteroids_data)>=target:
                     break
        if len(asteroids_data)>=target:
            break
    if len(asteroids_data)>=target:
        break
url = data["links"]["next"] # Pagination 
                     
```

```python
import mysql.connector as db # import packages 
```

```python
connection = db.connect(
                        host = "localhost",
                        user = "root",
                        password = "1234",
                        database = "nasa"
                       ) # connect
```

```python
connection
```

```python
cursor = connection.cursor() # create cursor
```

```python
# Create asteroids table to store general info

cursor.execute ("""create table asteroids (
                date_collected DATE,
                id INT,
                name TEXT,
                absolute_magnitude_h FLOAT,
                estimated_diameter_min_km FLOAT,
                estimated_diameter_max_km FLOAT,
                is_potentially_hazardous_asteroid BOOLEAN); """)

connection.commit()


```

```python
# Create Close_approach table to store asteroid's approach to Earth

cursor.execute ("""create table close_approach (
                neo_ref_id INT,
                close_approach_date DATE,
                relative_velocity_kmph FLOAT,
                astronomical FLOAT,
                miss_distance_km FLOAT,
                miss_distance_lunar FLOAT,
                orbiting_body TEXT); """)

connection.commit()

print("Tables created successfully.")
```

```python
#insert data to table

for asteroid in asteroids_data:
    # Insert into asteroids table
    asteroid_query = """INSERT INTO asteroids (date_collected, id, name, absolute_magnitude_h, estimated_diameter_min_km, estimated_diameter_max_km, is_potentially_hazardous_asteroid)
                        VALUES (%s, %s, %s, %s, %s, %s, %s)"""
    values = (
         asteroid['date_collected'],
         asteroid['id'],
         asteroid['name'],
         asteroid['absolute_magnitude_h'],
         asteroid['estimated_diameter_min_km'],
         asteroid['estimated_diameter_max_km'],
         asteroid['is_potentially_hazardous_asteroid'])
    cursor.execute(asteroid_query, values)

connection.commit()
print("Inserted asteroids data")
```

```python
#insert approach data to table 

for approach in asteroids_data:
    # Insert into close_approach table
    approach_query = """INSERT INTO close_approach (neo_ref_id, close_approach_date, relative_velocity_kmph, astronomical, miss_distance_km, miss_distance_lunar, orbiting_body)
                        VALUES (%s, %s, %s, %s, %s, %s, %s)"""
    approach_values = (
         approach['neo_ref_id'],
         approach['close_approach_date'],
         approach['relative_velocity_kmph'],
         approach['astronomical'],
         approach['miss_distance_km'],
         approach['miss_distance_lunar'],
         approach['orbiting_body'])
    cursor.execute(approach_query, approach_values)

connection.commit()
print("Inserted close_approach data")
```
